```python
#CLUSTERING METHOD 1 - KMEANS
#Visualization of some data points
import matplotlib.pyplot as plt
breast_cancer = datasets.load_breast_cancer()
x = breast_cancer.data
y = breast_cancer.target
feature1 = 0
feature2 = 1
plt.scatter(x[:, feature1], x[:, feature2])
plt.show()
```

```python
#Elbow method is used to get the best value of K
from sklearn.cluster import KMeans
import numpy as np
data = np.column_stack((x, y))
inertias = []

for j in range(1,11):
    kmeans = KMeans(n_clusters=j)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
##CLUSTERING METHOD 1 - KMEANS MODEL
from sklearn.cluster import KMeans
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
breast_cancer = datasets.load_breast_cancer()
x = breast_cancer.data
y = breast_cancer.target
feature_names = breast_cancer.feature_names
target_names = breast_cancer.target_names
#The optimal no of clusters from the Elbow method above is 2
k = 2
kmeans = KMeans(n_clusters= k)
kmeans. fit(x)
labels = kmeans.labels_
cluster_centres = kmeans.cluster_centers_
print('clusters:', labels)
print('cluster_centres:', cluster_centres)
from numpy.random import rand
plt.scatter(cluster_centres[:, 0] , cluster_centres[:, 1])
colors = ['r','y']
markers = ['o', 'd']
for j in range(k):
  plt.scatter(x[labels == j, 0], x[labels == j, 1], s=50, c=colors[j], marker=markers[j], label=f'Cluster {j+1}')
plt.scatter(cluster_centres[:, 0], cluster_centres[:, 1], s=200, marker='*', c='black', label='Centroids')
plt.xlabel(feature_names[0])
plt.ylabel(feature_names[1])

plt.title('KMEANS CLUSTER SCATTERED GRAPH')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn import datasets, svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split

# IMPORT DATASET
breast_cancer = datasets.load_breast_cancer()
x = breast_cancer.data
y = breast_cancer.target
class_names = breast_cancer.target_names

# Split the data into a training set and a test set
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)

# Run classifiers, using models to see the impact on the results
classifier = svm.SVC(kernel="linear", C=0.01).fit(x_train, y_train)
classifier = rfc = RandomForestClassifier(n_estimators=100, random_state=0).fit(x_train, y_train)
classifier = knn= KNeighborsClassifier (n_neighbors=2).fit(x_train, y_train)

np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
titles_options = [
    ("Confusion matrix, without normalization", None),
    ("Normalized confusion matrix", "true"),
]
for title, normalize in titles_options:
    disp = ConfusionMatrixDisplay.from_estimator(
        classifier,
        x_test,
        y_test,
        display_labels=class_names,
        cmap=plt.cm.Blues,
        normalize=normalize,
    )
    disp.ax_.set_title(title)

    print(title)
    print(disp.confusion_matrix)

plt.show()
```
